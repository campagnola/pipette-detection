```python
import numpy as np
import matplotlib.pyplot as plt

from make_training_data import make_training_data, PipetteTemplate
```

```python
template = PipetteTemplate('yip_2019_template.npz')
```

```python
shape = (10, 5)
fig, ax = plt.subplots(shape[0], shape[1], figsize=(shape[1]*3, shape[0]*3))

for j in range(shape[1]):
    difficulty = j/(shape[1]-1)
    ax[0,j].set_title(f"difficulty={difficulty:0.2f}")
    for i in range(shape[0]):
        image, pip_pos = make_training_data(difficulty=difficulty, shape=(500, 500), template=template)
        ax[i,j].imshow(image, cmap='gray')
        ax[i,j].scatter(pip_pos[2], pip_pos[1], color='red')
```
